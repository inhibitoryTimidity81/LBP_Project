```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import scipy.io
```

```python
import os
for dirname, _, filenames in os.walk('dataset'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Load the .mat file (replace 'dataset/B007_1_123.mat' with your file path)
mat_data = scipy.io.loadmat('dataset/B007_1_123.mat')
```

```python
print(mat_data.keys())
```

```python
df_DE = pd.DataFrame(mat_data['X123_DE_time'])
df_FE = pd.DataFrame(mat_data['X123_FE_time'])
```

```python
df_DE.shape
```

```python
df_FE.shape
```

```python
#plot the distribution of the DE features for first 20000 points
plt.figure(figsize=(5, 4))
plt.plot(df_DE.iloc[:50000, 0], color='orange')
plt.title('DE Feature over Time (First 50000 Points)')   
plt.xlabel('Time Frame')
plt.ylabel('DE Acceleration')
plt.show()
```

```python
#plot the distribution of the FE features for first 20000 points
plt.figure(figsize=(10, 6))
plt.plot(df_FE.iloc[:50000, 0])
plt.title('FE Feature over Time (First 50000 Points)')   
plt.xlabel('Time Frame')
plt.ylabel('FE Acceleration')
plt.show()
```
